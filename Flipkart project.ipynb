# Boxplot

Boxplot gives a nice summary of the data. It helps in understanding our distribution better.

```python
sns.boxplot(x='sex', y='total_bill', data=df1, palette='Set1')
```

```python
bp= sns.boxplot(x='sex', y='total_bill', data=df1, palette='Set1')
mean = df1.groupby(['sex'])['total_bill'].mean()
for i in bp.get_xticks():
    bp.text(i, mean[i], mean[i], fontsize=10, color='black')
```

# Voilin plot

Violin plot is a better chart than boxplot as it gives a much broader understanding of the distribution. It resembles a violin and dense areas point the more distribution of data otherwise hidden by box plots

```python
sns.violinplot(x="sex", y="total_bill",data=df1,palette='Set1')
```

```python
sns.violinplot(x="sex", y="total_bill",data=df1,palette='Set1', hue="sex")
```

# PairGrid

PairGrid in Seaborn is used to create a grid of subplots where each variable is plotted against every other variable in a dataset

```python
sns.PairGrid(df1).map(plt.scatter)
```

```python
m= sns.PairGrid(df1)
m.map_diag(plt.hist)
m.map_upper(plt.scatter)
m.map_lower(sns.kdeplot)
```

# FaceGrid

FacetGrid in Seaborn is used to create a grid of plots based on subsets of data. It allows you to visualize the relationship between variables across different subsets of the data, making it useful for comparing patterns and trends.(100+ charts)

```python
f = sns.FacetGrid(df1, col="time", row="day")
```

```python
f = sns.FacetGrid(df1, col="time",  row="day")
f = f.map(plt.hist, "size")
```

```python
g = sns.FacetGrid(df1, col="tip",  row="time",hue='sex')
g = g.map(plt.scatter, "time", "total_bill").add_legend()
```

# Implot

It to understand the correlation between different features in your dataset, assess the goodness of fit of a regression model, and identify outliers or nonlinear patterns in the data.

```python
sns.lmplot(x='total_bill',y='tip', hue="sex", row="time", col="size", data=df1)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='time', data=df1, palette='deep')
```

```python
sns.set_context('poster', font_scale=2)
sns.countplot(x='time', data=df1, palette="coolwarm")
```

# Grid Type Plot

Grid-type plots, such as FacetGrid, PairGrid, and JointGrid in Seaborn, are used for organizing and visualizing data relationships in a structured grid layout. They allow for the creation of multiple plots arranged in rows and columns.

```python
sns.lmplot(x='total_bill',y='tip',aspect=4,data=df1)
```
